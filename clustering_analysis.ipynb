```
from sklearn.datasets.samples_generator import make_blobs
```

```
x,y=make_blobs(n_samples=300,centers=4,cluster_std=0.60,random_state=0)
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(x[:,0],x[:,1],s=50)
```

```
from sklearn.cluster import KMeans
```

```
kmeans=KMeans(n_clusters=64)
```

```
kmeans.fit(x)
```

```
y_means=kmeans.predict(x)
```

```
plt.scatter(x[:,0],x[:,1],s=50,c=y_means,cmap='viridis')
```

```
centers =kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black',s=200,alpha=0.5)
```

```
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

```
x,y=make_blobs(random_state=42)
agg=AgglomerativeClustering(n_clusters=3)
k=agg.fit_predict(x)
mglearn.discrete_scatter(x[:,0],x[:,1],k)
```

```
mglearn.plots.plot_agglomerative()
```

```
from scipy.cluster.hierarchy import dendrogram, ward
```

```
x,y=make_blobs(random_state=0,n_samples=12)
linkange_array=ward(x)
```

```
dendrogram(linkange_array)
```

```
from sklearn.cluster import DBSCAN
```

```
from sklearn.datasets import make_moons
```

```
x,y=make_moons(random_state=42,n_samples=250,noise=0.05)
```

```
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
x_scaled=scaler.transform(x)
```

```
dbscan=DBSCAN()
```

```
kumeler=dbscan.fit_predict(x_scaled)
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=kumeler,cmap=mglearn.cm2,s=60)
```
